```python
#import packages
import pandas as pd
import numpy as np

#read in files
space_train = pd.read_csv('train.csv')
space_test = pd.read_csv('test.csv')

#add transported column to test data
space_test['Transported'] = [np.nan] * 4277

#join dataframes together for one large dataset
space = pd.concat([space_train, space_test], ignore_index = True)

#total length of dataframe
length = space.shape[0]

space.tail(20)
```

```python
#splitting PassengerId into group and passenger number

group = []
psngr = []

for r in range(length):
    x  = space['PassengerId'][r].split('_')
    group.append(x[0])
    psngr.append(x[1])

space['Group'] = group
space['Passenger'] = psngr

#splitting Name into first and last name

first = []
last = []

for r in range(length):
    x  = space['Name'][r]
    if type(x) == str:
        y = x.split()
        first.append(y[0])
        last.append(y[1])
    else:
        first.append(np.nan)
        last.append(np.nan)

space['FirstName'] = first
space['LastName'] = last

#splitting Cabin into deck, number and side

deck = []
num = []
side = []

for r in range(length):
    x  = space['Cabin'][r]
    if type(x) == str:
        y = x.split('/')
        deck.append(x[0])
        num.append(x[2])
        side.append(x[4])
    else:
        deck.append(np.nan)
        num.append(np.nan)
        side.append(np.nan)

space['Deck'] = deck
space['Number'] = num
space['Side'] = side


space.head(20)
```

```python
#determining if passengers are in a group or not

ingroup = ['False'] * length

for first in range(length):
    for second in range(length):
        if ingroup[first] == 'True':
            break
        if space['Group'][first] == space['Group'][second] and first != second:
            ingroup[first] = 'True'
            ingroup[second] = 'True'
            break
            
space['InGroup'] = ingroup

space.head(20)
```

```python
#determining if members within the same group have the same last name

infamily = ['False'] * length

for first in range(length):
    if ingroup[first] == 'True':
        for second in range(length):
            if infamily[first] == 'True':
                break
            if ingroup[second] == 'True':
                if space['Group'][first] == space['Group'][second] and first != second and space['LastName'][first] == space['LastName'][second]:
                    infamily[first] = 'True'
                    infamily[second] = 'True'
                    break
            
space['InFamily'] = infamily

space.head(20)
```

```python
#creating dataframe of relevant information

space = space.drop(columns = ['Cabin', 'Name'])

```
