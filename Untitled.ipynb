```python
#import packages
import pandas as pd
import numpy as np

data = pd.read_csv('myspacedata.csv')
        
length = data.shape[0]

data = data.set_index('Index')

data
```

```python
data.mode(axis = 0)
```

```python
data.mean(axis = 0)
```

```python
data['HomePlanet'] = data['HomePlanet'].fillna('Earth')
data['CryoSleep'] = data['CryoSleep'].fillna('False')
data['Destination'] = data['Destination'].fillna('TRAPPIST-1e')
data['Age'] = data['Age'].fillna(29.0)
data['VIP'] = data['VIP'].fillna('False')
data['RoomService'] = data['RoomService'].fillna(223.0)
data['FoodCourt'] = data['FoodCourt'].fillna(452.0)
data['ShoppingMall'] = data['ShoppingMall'].fillna(175.0)
data['Spa'] = data['Spa'].fillna(308.0)
data['VRDeck'] = data['VRDeck'].fillna(307.0)
data['InGroup'] = data['InGroup'].fillna('False')
```

```python
data = data.drop(columns = ['FirstName', 'LastName', 'Deck', 'Number', 'Side'])
```
